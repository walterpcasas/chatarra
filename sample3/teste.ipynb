```python
import numpy as np
from scipy.spatial import distance

matrix1 = np.array([[0, 1, 0], [1, 0, 1], [0, 1, 0]])
matrix2 = np.array([[1, 0, 1], [0, 1, 0], [1, 0, 1]])

# Obtener las coordenadas de los puntos en cada matriz
coords1 = np.column_stack(np.where(matrix1 == 1))
coords2 = np.column_stack(np.where(matrix2 == 1))

# Obtener la distancia euclidiana entre cada punto en una matriz y un punto específico en la otra matriz
point1 = coords1[0,:]
point2 = coords2[0,:]
distances = [distance.euclidean(point1, p) for p in coords2]

print(distances)
```
